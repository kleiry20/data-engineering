# Problem 1:
# You've decided to keep a journal. You have your first three entries ready to save:
# Write the entries to a file called journal.txt, one per line. Then read the file back and print its contents.

```python
# TODO: Write the entries to journal.txt, one per line, then read and print the contents
entries = [
    "Day 1: Started learning Python",
    "Day 2: Went for a run",
    "Day 3: Built my first Python script",
]

# WRITE THE CONTENT (write mode)
file = open("../file-handling/journal.txt", "w") # -> file is in a string format
file.write("\n".join(entries))
file.close()

# READ THE CONTENT (read mode)
file = open("../file-handling/journal.txt", "r")
contents = file.read()
file.close()

# PRINT THE CONTENT:
print(contents)
```

# Problem 2: 
# A few more days have passed and you have new entries to add to your journal:
# Append these entries to journal.txt, one per line. Then read the file back and print all its contents to confirm the original entries are still there.

```python
# TODO: Append the entries to journal.txt, then read and print all contents
new_entries = [
    "Day 4: Had coffee with a friend",
    "Day 5: Completed a Python project",
    "Day 6: Read a book",
]

# APPEND THE ENTRIES (append mode)
file = open("../file-handling/journal.txt", "a")
file.write("\n" + "\n".join(new_entries))
file.close()

# READ THE CONTENT (read mode)
file = open("../file-handling/journal.txt", "r")
contents = file.read()
file.close()

# PRINT THE CONTENT
print(contents)
```

#  Problem 3: (Using Context Manager)
# You want to look back at your Python-related entries. Using context managers (with statements), read journal.txt and 
# create a new file called python_entries.txt containing only the entries that mention "Python". 
# Then read the new file back and print its contents.

```python
# TODO: Use context managers to read journal.txt, filter to entries mentioning "Python", write to python_entries.txt, 
# then read and print it
new_list = []
with open("journal.txt", "r") as file:
    contents = file.read()
    contents = contents.split("\n") # -> file is a big string, we need to split it into list, can be done within same variable
    for i in contents:
        if "Python" in i or "python" in i:
            new_list.append(i)

with open("../file-handling/python_entries.txt", "w") as file:
    file.write("\n".join(new_list))

with open("../file-handling/python_entries.txt", "r") as file:
    contents = file.read()
    print(contents)
```

Why is it IMPORTANT to close a Python file?
File handling steps involve: 
1. opening a file
2. file operations (read, write, append etc)
3. close the file

If we only create a new file and write to it, the new file will be empty.
That's because Python does not write changes to the file right away.
Instead, it holds the changes into a memory buffer. That's why calling close is important.
By calling close, we tell Python to push those changes from the buffer to the file.

Downsides:
- we can forget calling close
- unexpected errors may occur

Solution to the above downsides -> Context Managers
