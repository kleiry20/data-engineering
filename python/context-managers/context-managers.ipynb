Why is it IMPORTANT to close a Python file?
File handling steps involve: 
1. opening a file
2. file operations (read, write, append etc)
3. close the file

```python
# File Handling operations:
file = open("../context-managers/books.txt", "w")
file.write("The Hunger Games\n")
file.write("The Great Gatsby\n")
file.write("The Maze Runner\n")
file.close()
```

If we only create a new file and write to it, the new file will be empty.
That's because Python does not write changes to the file right away.
Instead, it holds the changes into a memory buffer. That's why calling close is important.
By calling close, we tell Python to push those changes from the buffer to the file.

Downsides:
- we can forget calling close
- unexpected errors may occur

Solution to the above downsides = Context Managers

```python
# Context Managers
with open("../context-managers/books.txt", "w") as file:
    file.write("The Hunger Games\n")
    file.write("The Great Gatsby\n")
    file.write("The Maze Runner\n")
```

Suppose there are 2 files: 
input.txt (input) and catalog.txt (output)

Task:
Read an input file (read data)
Perform a simple operation to sort the books (manipulate data)
Output the result into another file (write data)

To do this, we need to have both files open at the same time (like a nested with... as statement), we can do this by:
- nesting the two with blocks (as the code grows, nesting may get hard to follow)
- use one context manager for multiple files

```python
# Approach 1: nesting the with blocks
with open("../context-managers/input.txt", "r") as input_file:
    with open("../context-managers/catalog.txt", "w") as output_file:
        titles = input_file.read().split("\n")
        titles.sort()
        output_file.write("\n".join(titles))

print(titles)
```

```python
# Approach 2: 
with open("../context-managers/input.txt", "r") as input_file, open("../context-managers/catalog.txt", "w") as output_file:
        titles = input_file.read().split("\n")
        titles.sort()
        output_file.write("\n".join(titles))

print(titles)
```
